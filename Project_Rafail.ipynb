```python
# marging show and no show
dataset = pd.concat([showdf, noShowDf], axis=0)
len(dataset)
dataset.head(3)
```

```python
# Function for converting strig to numeric

def encode_target(df, target_column, column_name):
    """Add column to df with integers for the target.

    Args
    ----
    df -- pandas DataFrame.
    target_column -- column to map to int, producing
                     new Target column.

    Returns
    -------
    df_mod -- modified DataFrame.
    targets -- list of target names.
    """
    df_mod = df.copy()
    targets = df_mod[target_column].unique()
    map_to_int = {name: n for n, name in enumerate(targets)}
    df_mod[column_name] = df_mod[target_column].replace(map_to_int)

    return (df_mod, targets)
```

```python
dataset, _ = encode_target(dataset, "ClinicNM", "ClinicNM_1")
dataset, _ = encode_target(dataset, "AppointmentDTS", "AppointmentDTS_1")
dataset, _ = encode_target(dataset, "SexFLG", "SexFLG_1")
dataset, _ = encode_target(dataset, "InsuranceDSC", "InsuranceDSC_1")
```

```python
y = dataset.NoShowFLG
#need to drop non numeric data
X = dataset.drop(columns=['NoShowFLG','ClinicNM','AppointmentDTS','SexFLG',
                          'InsuranceDSC','AppointmentID','PatientID','CancelledLateFLG'
                         #,'EmailFLG','HypertensionFLG' # drop to test if it works better
                         ],axis=1)

# Rename columns
X = X.rename(columns={'ClinicNM_1':'ClinicNM','AppointmentDTS_1':'AppointmentDTS','SexFLG_1':'SexFLG','InsuranceDSC_1':'InsuranceDSC'})
```

```python
# setting up testing and training sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

print(len(X_train))
print(len(X_test))

sm = SMOTE(sampling_strategy='auto',random_state=42)
# make X-train balanced
X_train, y_train = sm.fit_sample(X_train, y_train)

print(len(X_train))
```

# Decision Trees

```python
# Decision Tree Features
'''
0 criterion: str {'gini','entropy'}
1 splitter :str {'best', 'random'} 
2 max_depth : int or None, optional (default=None)
3 min_samples_split : int, float, optional (default=2)
4 min_samples_leaf : int, float, optional (default=1)
5 min_weight_fraction_leaf : float, optional (default=0.)
6 max_features: int, float, str or None, optional (default=None)
7 random_state : int, RandomState instance or None, optional (default=None)
8 max_leaf_nodes : int or None, optional (default=None)
9 min_impurity_decrease : float, optional (default=0.)
'''

def create_decision_tree_clissifier_parameters():
    criterion = np.random.choice(['gini','entropy']) #np.random.randint(low=2, high=16)
    splitter = np.random.choice(['best', 'random'])
    max_depth = np.random.choice([None,np.random.randint(low=2, high=100)])
    min_samples_split = np.random.randint(low=2, high=100)
    min_samples_leaf =  np.random.choice([1,np.random.randint(low=2, high=100)])
    min_weight_fraction_leaf = np.random.uniform(low = 0.0, high= 0.05)
    max_features = np.random.uniform(low = 0.0001, high= 1.0)
    random_state = 99 #np.random.randint(low=1, high=100)
    max_leaf_nodes = np.random.randint(low=2, high=500)
    min_impurity_decrease = np.random.choice([0.0,np.random.uniform(low=0.0, high=.05)])
    
    return [criterion, splitter, max_depth, min_samples_split, min_samples_leaf,
            min_weight_fraction_leaf, max_features, random_state, max_leaf_nodes, 
            min_impurity_decrease]

sample = create_decision_tree_clissifier_parameters()
print(sample)
```

```python
dt = tree.DecisionTreeClassifier(criterion = sample[0] , splitter = sample[1], max_depth = sample[2], 
                                           min_samples_split = sample[3], min_samples_leaf = sample[4],
                                           min_weight_fraction_leaf = sample[5], max_features = sample[6], 
                                           random_state = sample[7], max_leaf_nodes = sample[8], 
                                           min_impurity_decrease = sample[9]) 
```

```python

clf = dt.fit(X_train, y_train)
clf_pred = clf.predict(X_test)
# Checking accuracy
print(accuracy_score(y_test, clf_pred))
# f1 score
print(f1_score(y_test, clf_pred))
# recall score    
print(recall_score(y_test, clf_pred))
plot_confusion_matrix(clf, X_test, y_test) 
plt.show()
print(clf.feature_importances_)
print(X.columns[np.argmax(clf.feature_importances_)])
```

```python
# epoch
epoch = 200
parameter_hist = []
accuracy_hist = []
maxacc = [-9999,0]
maxf1 =[-999,0]
maxRc  = [-999,0]
for i in range(epoch):
    sample = create_decision_tree_clissifier_parameters()
    parameter_hist.append(sample)
    dt = tree.DecisionTreeClassifier(criterion = sample[0] , splitter = sample[1], max_depth = sample[2], 
                                           min_samples_split = sample[3], min_samples_leaf = sample[4],
                                           min_weight_fraction_leaf = sample[5], max_features = sample[6], 
                                           random_state = sample[7], max_leaf_nodes = sample[8], 
                                           min_impurity_decrease = sample[9])
    clf = dt.fit(X_train, y_train)
    clf_pred = clf.predict(X_test)
    # Checking accuracy
    acc = accuracy_score(y_test, clf_pred)
    # f1 score
    f1_scr = f1_score(y_test, clf_pred)
    # recall score    
    recall_scr = recall_score(y_test, clf_pred)
    
    if acc > maxacc[0] :
        maxacc[0] = acc
        maxacc[1] = i
    if f1_scr > maxf1[0] :
        maxf1[0] = f1_scr
        maxf1[1] = i
    if recall_scr > maxRc[0] :
        maxRc[0] = recall_scr
        maxRc[1] = i
        
#     plot_confusion_matrix(clf, X_test, y_test) 
#     plt.show()
    print(X.columns[np.argmax(clf.feature_importances_)])
    
    
    accuracy_hist.append([acc,f1_scr,recall_scr])
    print("Acc ",acc," F1 ",f1_scr," Recal ",recall_scr)
```

```python
accuracy_hist =  np.asarray(accuracy_hist)
fg = plt.figure(figsize=(20,10))
print(type(accuracy_hist))
plt.plot(np.arange(epoch),accuracy_hist[:,0],label="Accuracy")
plt.plot(np.arange(epoch),accuracy_hist[:,1],label="F1 Score")
plt.plot(np.arange(epoch),accuracy_hist[:,2],label="Recall Score")
plt.xlabel("Epoch")
plt.ylabel("Scores")
plt.title("Epoch vs. Scores")
plt.legend()
plt.show()
```
